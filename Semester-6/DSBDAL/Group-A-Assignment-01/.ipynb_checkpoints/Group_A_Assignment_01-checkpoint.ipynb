```python
# filling null values
df = df.fillna(value = 0)
df
```

```python
df = pd.read_csv("iris.csv")
df = df.fillna(method='pad')
df
```

```python
# ffill fills NaN value with the previous value
df = pd.read_csv("iris.csv")
df = df.fillna(method='ffill')
df
```

```python
#bfill fills the NaN value with the next value
df = pd.read_csv("iris.csv")
df = df.fillna(method='bfill')
df
```

```python
# Filling null values with mean
df = pd.read_csv("iris.csv")
mean_value = df['sepal_length'].mean()
df['sepal_length'].fillna(value=mean_value, inplace=True)
df
```

```python
# Filling null values with median
df = pd.read_csv("iris.csv")
median_value = df['sepal_length'].median()
df['sepal_length'].fillna(value=median_value, inplace=True)
df
```

```python
# Filling null values with mode
df = pd.read_csv("iris.csv")
mode_value = df['sepal_length'].mode()[0]
df['sepal_length'].fillna(value=mode_value, inplace=True)
df
```

```python
# Data formatting and data normalization
df.dtypes
```

```python
#Changing the data type of the column
# Before changging the data type of the column, all null values must be handled properly
mean_value = df['sepal_length'].mean()
df['sepal_length'].fillna(value=mean_value, inplace=True)
df['sepal_length'] = df['sepal_length'].astype(int)

df.dtypes
```

```python
mean_value = df['sepal_length'].mean()
df['sepal_length'].fillna(value=mean_value, inplace=True)

df['sepal_length'] = df['sepal_length'].round(0).astype(int)
df.dtypes
```

```python
# Converting categorical variables into quantitive variables
df['species'].replace({'Iris-setosa':1, 'Iris-versicolor':2, 'Iris-virginica':3}, inplace=True)
df
```
