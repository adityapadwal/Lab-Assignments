```python
#Filling null values of Math_Score with mean of the column
df = pd.read_csv("Academic-Performance-Dataset.csv")
mean_value = df['Maths_Score'].mean().round()
df['Maths_Score'].fillna(value=mean_value, inplace=True)
df.head()
```

```python
#Filling null values of Reading_Score column with median
median_value = df['Reading_Score'].median()
df['Reading_Score'].fillna(value=median_value, inplace=True)
df.head()
```

```python
#Filling null values of Writing_Score with mode
mode_value = df['Writing_Score'].mode()[0]
df['Writing_Score'].fillna(value=mode_value, inplace=True)
df.head()
```

```python
# Replacing the missing values in Placement_Score column with the min value
df['Placement_Score'].fillna(value=df['Placement_Score'].min(), inplace=True)
df.head()
```

```python
#Replacing the missing values in Club_Join_Date with the max value
df['Club_Join_Date'].fillna(value=df['Club_Join_Date'].max(), inplace=True)
df.head()
```
